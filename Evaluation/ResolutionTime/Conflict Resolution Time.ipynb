```julia
spawnRateSet=[5,15,25,35,45,55,65]/100;
pe_tr_plot = Axis([
    Plots.Linear(spawnRateSet, [avgResTimeNOCAS[key] for key in spawnRateSet], legendentry="NOCAS", mark="*", style="mark options={fill=black}, black, very thick"), 
    Plots.Linear(spawnRateSet, [avgResTimeCorrectedSector[key] for key in spawnRateSet], legendentry="CorrectedSector", mark="*", style="mark options={fill=magenta}, magenta, very thick"), 
    Plots.Linear(spawnRateSet, [avgResTimeCorrectedClosest[key] for key in spawnRateSet], legendentry="CorrectedClosest", mark="*", style="mark options={fill=pink}, pink, very thick"), 
    Plots.Linear(spawnRateSet, [avgResTimeVICASMulti[key] for key in spawnRateSet], legendentry="VICASMulti", mark="*", style="mark options={fill=blue}, blue, very thick"),
    Plots.Linear(spawnRateSet, [avgResTimeVICASClosest[key] for key in spawnRateSet], legendentry="VICASClosest", mark="*", style="mark options={fill=red}, red, very thick"),
    Plots.Command("\\legend{}")
    ], 
    ymin=16,
    legendStyle="{at={(0.02,0.02)}, anchor=south west}",
    xlabel="Prob. of New Encounter per Time Step, Pe",
    ylabel="Time Steps to Resolve an Encounter, tr (s)",
    width="3.2in",
    height="3in"
)
```

```julia
PGFPlots.save("eff3.tex", pe_tr_plot);
```
